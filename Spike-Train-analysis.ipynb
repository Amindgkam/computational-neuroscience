```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Spike Generation
def generate_spike_train(duration, rate):
    time = np.arange(0, duration, 1)  # Time in ms
    print("Time",time)
    spikes = np.random.rand(len(time)) < (rate / 1000)  # Spike probability
    print("Spikes: ",spikes)
    print(spikes.shape)
    return spikes.astype(int)
```

```python
# Calculate the Peri-Stimulus Time Histogram (PSTH)
def calculate_psth(spike_train, bin_size=50):
    hist, _ = np.histogram(np.where(spike_train)[0], bins=np.arange(0, len(spike_train), bin_size))
    print(f"Hist: {hist}")
    return hist
```

```python
# Parameters
duration = 200 # Duration in ms
rate = 100 # Spike rate in Hz
```

```python
# Create spike train and PSTH
spike_train = generate_spike_train(duration, rate)
psth = calculate_psth(spike_train)
```

```python
# Plotting
plt.figure(figsize=(10, 5))
plt.subplot(2, 1, 1)
plt.plot(spike_train, marker='o', linestyle='None', label='Spike Train')
plt.title('Spike Train')
plt.xlabel('Time (ms)')
plt.ylabel('Spike (1=spike, 0=no spike)')
plt.subplot(2, 1, 2)
plt.bar(np.arange(len(psth)) * 50, psth, width=50)
plt.title('Peri-Stimulus Time Histogram (PSTH)')
plt.xlabel('Time (ms)')
plt.ylabel('Spike Count')
plt.tight_layout()
plt.show()
```
